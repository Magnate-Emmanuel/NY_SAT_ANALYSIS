### QUESTION 1

##### Define the vector (from the task) that will be used to test the code. It is always best practice to test codes this way before applying them on large datasets. You do not necessarily have to define it before the function. 

```python
arr = [1, 2, 3, 2, 1, 2, 3]
```

```python
def myFunc(arr):
    #predine the window of numbers to calculate the moving-max over
    window = 3  #putting this outside of the function will still work
    
    #Initializing an empty list to store the moving max
    #This is important as we need to organize the moving-max back into a vector
    moving_max = []
   
    #Getting the number of iterations/looping to be done based on the window
    no_iterations = len(arr) - window + 1
    
    #Loop through the vector
    #Insofar as the range defined has not been exhausted, the iteration continues
    for i in range(no_iterations):
        window_i = arr[i: i+window] #Getting the values of the vector of the moving window
        
        max_value = max(window_i) # Calculate the max value of the current window
        moving_max.append(max_value) # Append the max value of the current window to moving_max list.
    return moving_max #return the moving_max
```

```python
print(myFunc(arr))   #Everything works fine
```

### QUESTION 2

##### I issue a GET request on the url which gets the text data. Clean the data a bit and then call the myFunc on it. Finally, I calculate the sum of the vector of moving max generated

```python
#Define the url to get the data from as given in the task
url = 'https://www.angio.net/pi/digits/100000.txt'

# The get() method sends a GET request to the specified url
result = requests.get(url).content #The content is assigned to the variable result

#Decode the data
data = result.decode('utf-8')  #This converts bytes into string making it easier to index
data = ''.join(data.split('.')) 
values = []
for i in data:
    values.append(int(i)) #After the loop i done, there will be a vector like we had in question one
print(myFunc(values))
```

```python
summ = sum(myFunc(values))
print(summ)
```

## QUESTION 3

```python
def schools(a):
    if 'HIGH SCHOOL' in a.upper() and 'ACADEMY' not in a.upper(): #It is important to add "and 'ACADEMY' not in a.upper()" else "the both" category will not show
       return 'High School'
    elif 'ACADEMY' in a.upper() and 'HIGH SCHOOL' not in a.upper(): #same is true here
       return 'Academy'
    elif 'HIGH SCHOOL' in a.upper() and 'ACADEMY' in a.upper():
      return 'Both' #Both only work when you do not allow schools with both keywords to be assigned to either "high" or "academy" already
    else:
        return 'Other'
```

```python
sat['category'] = sat['SCHOOL NAME'].apply(schools).values
sat.columns  #shows the column names
```

```python
sns.scatterplot('SAT Math Avg. Score','SAT Critical Reading Avg. Score', data = sat, hue='category').set(title= 'SAT Score by School Name', xlabel= 'Math', ylabel= 'Reading')
```
